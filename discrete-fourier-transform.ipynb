```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function to illustrate multiple periods of e^(i*pi)
def plot_multiple_periods_dft(N=50, periods=3):
    """
    Illustrates multiple periods of e^(i*pi) in the context of the DFT.
    
    Parameters:
    N -- Number of points per period (default 50)
    periods -- Number of periods to display (default 3)
    """
    # Generate time indices to cover multiple periods
    n = np.arange(N * periods)
    
    # Compute the complex exponential values (e^(-2*pi*i*k/N)) for k=1
    k = 1  # Frequency index
    exponent = -2 * np.pi * k * n / N  # Exponential term
    real_part = np.cos(exponent)  # Real part of the exponential
    imag_part = np.sin(exponent)  # Imaginary part of the exponential

    # Highlight the e^(i*pi) points (every half-period)
    pi_indices = np.where(np.abs((exponent / np.pi) % 2 - 1) < 1e-6)[0]
    highlight_real = real_part[pi_indices]
    highlight_imag = imag_part[pi_indices]

    # Plot in 3D
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Plot the full DFT points in 3D
    ax.scatter(n, real_part, imag_part, color='b', label='e^(-2πi * k * n / N)')
    ax.plot(n, real_part, imag_part, color='blue', alpha=0.6)

    # Highlight all e^(i*pi) points
    ax.scatter(
        n[pi_indices], highlight_real, highlight_imag,
        color='red', s=100, label='$e^{i\\pi} = -1$'
    )

    # Set labels and title
    ax.set_xlabel('Time Index (n)', fontsize=12)
    ax.set_ylabel('Real Part', fontsize=12)
    ax.set_zlabel('Imaginary Part', fontsize=12)
    ax.set_title(r'Multiple Periods of $e^{i\pi}$ in the Context of DFT', fontsize=14)  #raw string
        
    ax.legend()

    # Customize the viewing angle
    ax.view_init(elev=30, azim=40)
    plt.show()

# Call the function with more periods (e.g., periods=5)
#lot_multiple_periods_dft(N=20, periods=3) -- static plot

# Create an interactive widget for the plot
interactive_plot = interactive(
    plot_multiple_periods_dft,
    N=widgets.IntSlider(min=10, max=100, step=10, value=50, description="Points per Period (N)"),
    periods=widgets.IntSlider(min=1, max=10, step=1, value=3, description="Periods")
)

# Display the interactive widget
interactive_plot
```


Understanding $e^{i\pi} = -1$

**Rotational Interpretation (Unit Circle):**

$ e^{i\pi} $ represents a half-turn (180°) rotation on the complex plane.
Starting at 1 (the real number on the unit circle), a rotation by $ \pi $ radians lands you at -1.

**Implications of -1:**

- In physical systems, -1 often indicates a phase inversion or opposite direction.
- In signals, it represents a complete half-cycle shift, where a waveform flips polarity.


## Real-Life Applications of $e^{i\pi}$

### 1. Audio and Sound Waves

**Spherical Sound Waves:**

- Sound waves propagate as spherical waves (e.g., from a point source).
- These waves are oscillatory and can be represented using complex exponentials like $e^{i\theta}$, where the real and imaginary parts describe the wave's oscillation in space and time.
- When $e^{i\pi} = -1$, it corresponds to an oscillation completing half a wavelength, meaning the sound wave is at its negative peak.

**Phase Inversion:**

- In audio processing, $-1$ represents a phase shift of $\pi$ radians (180°), flipping the waveform. This is commonly used in:
    - Noise cancellation (e.g., active noise-canceling headphones).
    - Mixing and processing sound waves to create interference patterns.

**Fourier Transform in Audio:**

- Audio signals are decomposed into sinusoidal components using Fourier transforms, which rely on $e^{i\theta}$.
- Each frequency in the audio spectrum is a sum of oscillations, and $e^{i\pi}$ describes how the phase shifts affect the overall sound.

### 2. Memory Systems (Signal Propagation)

**Implications in Memory Systems:**

- Memory systems (e.g., in RAM or communication systems) rely on signal timing and synchronization.
- Oscillations governed by $e^{i\theta}$ describe how electrical signals propagate in circuits.
- A phase shift of $\pi$ (as in $e^{i\pi} = -1$) could indicate a complete inversion, which is crucial for timing, clock synchronization, or error correction.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interactive

# Define a function to generate and plot the sinusoidal components
def plot_dft_component(k=1, N=10):
    """
    Plots the sinusoidal component of the discrete Fourier transform (DFT).
    
    Parameters:
    k -- Frequency index (1 <= k <= N-1)
    N -- Total number of points in the DFT
    """
    # Generate time points
    n = np.arange(N)
    
    # Compute the real and imaginary parts of the exponential e^(-2*pi*i*k/N)
    real_part = np.cos(-2 * np.pi * k * n / N)
    imag_part = np.sin(-2 * np.pi * k * n / N)
    
    # Create 3D plot
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    
    # Plot the sinusoidal components
    ax.plot(n, real_part, imag_part, label=f"DFT Component (k={k}, N={N})")
    ax.scatter(n, real_part, imag_part, color='r')  # Mark the points
    
    # Set labels and title
    ax.set_xlabel("Time Index (n)")
    ax.set_ylabel("Real Part")
    ax.set_zlabel("Imaginary Part")
    ax.set_title(f"3D Visualization of DFT Component (k={k})")
    ax.legend()
    
    # Adjust viewing angle
    ax.view_init(elev=20, azim=30)
    
    # Show the plot
    plt.show()

# Create an interactive widget
interactive_plot = interactive(plot_dft_component, k=(1, 20, 1), N=(5, 50, 1))
interactive_plot
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interactive, widgets

# Advanced function to plot sinusoidal components of the DFT
def plot_dft_component(k=1, N=10, elevation=20, azimuth=30):
    """
    Advanced 3D visualization of a discrete Fourier transform (DFT) component.

    Parameters:
    k -- Frequency index (1 <= k <= N-1)
    N -- Total number of points in the DFT (N > 0)
    elevation -- Elevation angle for the 3D plot (degrees)
    azimuth -- Azimuth angle for the 3D plot (degrees)
    """
    # Generate time indices
    n = np.arange(N)

    # Compute the real and imaginary parts of e^(-2*pi*i*k/N)
    real_part = np.cos(-2 * np.pi * k * n / N)
    imag_part = np.sin(-2 * np.pi * k * n / N)
    
    # Calculate magnitude for color mapping
    magnitude = np.sqrt(real_part**2 + imag_part**2)  # pythagorean theorem

    # Create the 3D figure
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')

    # Plot the real and imaginary components as a 3D curve
    scatter = ax.scatter(n, real_part, imag_part, c=magnitude, cmap='viridis', s=50, label="Magnitude")
    curve = ax.plot(n, real_part, imag_part, color='blue', alpha=0.6, label=f"DFT Component (k={k}, N={N})")

    # Add color bar to indicate magnitude
    cbar = plt.colorbar(scatter, ax=ax, shrink=0.6, aspect=10, pad=0.1)
    cbar.set_label("Magnitude (|Real + Imag|)", fontsize=10)

    # Set labels, title, and legends
    ax.set_xlabel("Time Index (n)", fontsize=12)
    ax.set_ylabel("Real Part", fontsize=12)
    ax.set_zlabel("Imaginary Part", fontsize=12)
    ax.set_title(f"3D DFT Visualization (k={k}, N={N})", fontsize=14)
    ax.legend()

    # Set custom viewing angles
    ax.view_init(elev=elevation, azim=azimuth)

    # Show the final plot
    plt.show()

# Create an interactive widget for the advanced plot
interactive_plot = interactive(
    plot_dft_component,   #for every change in the widget, the function plot_dft_component is called
    k=widgets.IntSlider(min=1, max=20, step=1, value=1, description="Frequency (k)"),
    N=widgets.IntSlider(min=5, max=50, step=1, value=10, description="Points (N)"),
    elevation=widgets.IntSlider(min=0, max=90, step=5, value=20, description="Elevation"),
    azimuth=widgets.IntSlider(min=0, max=360, step=10, value=30, description="Azimuth")
)

# Display the interactive widget
interactive_plot
```
