```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Define the trigonometric functions
def plot_trig_functions(amp=1.0, omega=1.0):
    x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)  #domain with 1000 points

    sin = amp * np.sin(omega * x)  #omega = 2*pi/T   The frequency is cycles/sec. The period, T, is defined as the time it takes to complete one cycle.  
    cos = amp * np.cos(omega * x)
    tan = amp * np.tan(omega * x)
    cot = amp / np.tan(omega * x)
    sec = amp / np.cos(omega * x)
    csc = amp / np.sin(omega * x)

    plt.figure(figsize=(12, 8))

    # Plot sin(x)
    plt.plot(x, sin, label='sin(x)', color='blue')

    # Plot cos(x)
    plt.plot(x, cos, label='cos(x)', color='red')

    # Plot tan(x) with limited y-axis for visibility
    plt.plot(x, tan, label='tan(x)', color='green')
    plt.ylim(-10, 10)

    # Plot cot(x), excluding undefined points
    cot[np.isinf(cot)] = np.nan   #np.isinf() is a boolean function used to check if the elements are infinite
    plt.plot(x, cot, label='cot(x)', color='black')

    # Plot sec(x), excluding undefined points
    sec[np.isinf(sec)] = np.nan
    plt.plot(x, sec, label='sec(x)', color='black')

    # Plot csc(x), excluding undefined points
    csc[np.isinf(csc)] = np.nan  
    plt.plot(x, csc, label='csc(x)', color='black')

    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
    plt.axvline(0, color='black', linewidth=0.5, linestyle='--')

    plt.title('Trigonometric Functions')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()

# Create sliders for amp and omega
amp_slider = FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='amp:')
omega_slider = FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='omega:')

# Use interact to link sliders with the plot function
interact(plot_trig_functions, amp=amp_slider, omega=omega_slider)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Define the trigonometric functions
def plot_trig_functions(amp=1.0, omega=1.0):
    x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

    y_sin = amp * np.sin(omega * x)
    y_cos = amp * np.cos(omega * x)
    y_tan = amp * np.tan(omega * x)
    y_cot = amp / np.tan(omega * x)
    y_sec = amp / np.cos(omega * x)
    y_csc = amp / np.sin(omega * x)

    fig, axes = plt.subplots(3, 2, figsize=(14, 10))
    axes = axes.flatten()  #converts the 3x2 axes array to a 1D array

    # Plot sin(x)
    axes[0].plot(x, y_sin, label='sin(x)', color='blue')
    axes[0].axhline(0, color='black', linewidth=0.5, linestyle='--')
    axes[0].axvline(0, color='black', linewidth=0.5, linestyle='--')
    axes[0].set_title('sin(x)')
    axes[0].grid(True)

    # Plot cos(x)
    axes[1].plot(x, y_cos, label='cos(x)', color='orange')
    axes[1].axhline(0, color='black', linewidth=0.5, linestyle='--')
    axes[1].axvline(0, color='black', linewidth=0.5, linestyle='--')
    axes[1].set_title('cos(x)')
    axes[1].grid(True)

    # Plot tan(x) with limited y-axis for visibility
    axes[2].plot(x, y_tan, label='tan(x)', color='green')
    axes[2].axhline(0, color='black', linewidth=0.5, linestyle='--')
    axes[2].axvline(0, color='black', linewidth=0.5, linestyle='--')
    axes[2].set_ylim(-10, 10)
    axes[2].set_title('tan(x)')
    axes[2].grid(True)

    # Plot cot(x), excluding undefined points
    y_cot[np.isinf(y_cot)] = np.nan   #np.isinf() is a boolean function used to check if the elements are infinite
    axes[3].plot(x, y_cot, label='cot(x)', color='red')
    axes[3].axhline(0, color='black', linewidth=0.5, linestyle='--')
    axes[3].axvline(0, color='black', linewidth=0.5, linestyle='--')
    axes[3].set_ylim(-10, 10)
    axes[3].set_title('cot(x)')
    axes[3].grid(True)

    # Plot sec(x), excluding undefined points
    y_sec[np.isinf(y_sec)] = np.nan
    axes[4].plot(x, y_sec, label='sec(x)', color='purple')
    axes[4].axhline(0, color='black', linewidth=0.5, linestyle='--')
    axes[4].axvline(0, color='black', linewidth=0.5, linestyle='--')
    axes[4].set_ylim(-10, 10)
    axes[4].set_title('sec(x)')
    axes[4].grid(True)

    # Plot csc(x), excluding undefined points
    y_csc[np.isinf(y_csc)] = np.nan
    axes[5].plot(x, y_csc, label='csc(x)', color='brown')
    axes[5].axhline(0, color='black', linewidth=0.5, linestyle='--')
    axes[5].axvline(0, color='black', linewidth=0.5, linestyle='--')
    axes[5].set_ylim(-10, 10)
    axes[5].set_title('csc(x)')
    axes[5].grid(True)

    plt.tight_layout()
    plt.show()

# Create sliders for amp and omega
amp_slider = FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='amp:')
omega_slider = FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='omega:')

# Use interact to link sliders with the plot function
interact(plot_trig_functions, amp=amp_slider, omega=omega_slider)
```
