# wave playground

double-slit interference, but interactive. slide the slit separation and width and watch the fringes breathe. two effects multiplied: the fast cosine fringes from the two slits, under the slow $\mathrm{sinc}^2$ envelope from each slit's finite width.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

wavelength = 500e-9; L = 1.0
ys = np.linspace(-0.03, 0.03, 2000)
s = np.sin(np.arctan(ys/L))

def pattern(d_um=100.0, a_um=20.0):
    d, a = d_um*1e-6, a_um*1e-6
    env = (np.sinc(np.pi*a*s/wavelength/np.pi))**2
    fr = (np.cos(np.pi*d*s/wavelength))**2
    I = env*fr
    plt.figure(figsize=(9,3.5))
    plt.plot(ys*1000, I, color='tab:green')
    plt.plot(ys*1000, env, 'k--', lw=0.7)
    plt.xlabel('screen position (mm)'); plt.title(f'd={d_um:.0f}um  a={a_um:.0f}um')
    plt.show()

interact(pattern,
         d_um=FloatSlider(min=40, max=200, step=10, value=100),
         a_um=FloatSlider(min=10, max=40, step=5, value=20));
```

as a gif: opening up the slit separation pulls the fringes closer together (wider slits = more, finer fringes). this is the plot i put in the readme.

```python
from matplotlib.animation import FuncAnimation, PillowWriter
import os
os.makedirs('media', exist_ok=True)

a = 20e-6
env = (np.sinc(np.pi*a*s/wavelength/np.pi))**2
ds = np.linspace(60e-6, 220e-6, 60)
fig, ax = plt.subplots(figsize=(9,3.5))
line, = ax.plot([], [], color='tab:green')
ax.plot(ys*1000, env, 'k--', lw=0.7)
ax.set_xlim(-30, 30); ax.set_ylim(0, 1.05); ax.set_xlabel('screen position (mm)')
def update(i):
    I = env*(np.cos(np.pi*ds[i]*s/wavelength))**2
    line.set_data(ys*1000, I)
    ax.set_title(f'slit separation d = {ds[i]*1e6:.0f} um')
    return [line]
anim = FuncAnimation(fig, update, frames=len(ds), blit=False, interval=80)
anim.save('media/wave_playground.gif', writer=PillowWriter(fps=15))
plt.close(fig)
print('saved media/wave_playground.gif')
```
